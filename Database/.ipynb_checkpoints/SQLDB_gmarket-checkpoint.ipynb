```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import requests
from bs4 import BeautifulSoup
```

```python
url ='http://corners.gmarket.co.kr/Bestsellers'
```

```python
res = requests.get(url)
soup = BeautifulSoup(res.text, 'html.parser')
```

```python
# Main category 

path = '#categoryTabG > li> a' #id 는 한곳에서만 사용
lis = soup.select(path)

for li in lis:
#     print(li.text)
    print(li['href'], li.text)
    
```

```python
# 서브 카테고리

# http://corners.gmarket.co.kr/Bestsellers?viewType=G&groupCode=G01

path = '#categoryTabG > li> a' 
lis = soup.select(path)

main_url = 'http://corners.gmarket.co.kr'
for li in lis:
    main_cate_url = main_url + li['href']
    main_cate_title = li.text
    print(main_cate_url,main_cate_title)
    break
    
```
