```python
# country 

sql = """
select * from country;
"""
df_country = pd.read_sql(sql,db)
df_country.head(3)
```

```python
# city

sql = """
select * from city;
"""
df_city = pd.read_sql(sql,db)
df_city.head(3)
```

```python
# 합치기 위해 열이름 바꿔주기
df_city.rename(columns={'CountryCode':'Code'}, inplace = True)
df_city.head(3)
```

```python
# 두 데이터프레임 합치기 (key = 'Code')

df_total = pd.merge(df_country,df_city, on ="Code")
df_total.head(3)
```

```python
# (country name, city name)형식으로 출력하려 했으나 합병하면서 열의 이름이 바뀜

df_republic = df_total[df_total['GovernmentForm'].str.contains('Republic')]
df_republic.columns # Name_x 가 country name, Name_y 가 city name
```

```python
# Name_x 를 Country_Name, Name_y를 City_Name 로 바꿔줌
df_republic.rename(columns={'Name_x': 'Country_Name', 'Name_y': 'City_Name'}, inplace= True)
df_republic.head(3)
```

```python
# (country name, city name)형식으로 출력

df_republic[['Country_Name','City_Name','GovernmentForm']].sample(10)
```

```python
#sql 사용

# 1.'GovernmentForm'에 'Republic'이라는 텍스트가 포함된 국가를 출력
# 2. (country name, city name)형식으로 묶어서 출력

sql = """
select co.name as 'country', ci.name as 'city', governmentform from country co inner join city ci
    on co.code = ci.countrycode
    having governmentform like '%republic';
"""
# sql = 'select * from usertbl inner join buytbl 
#  on usertbl.userid=buytbl.userid'
# pd.read_sql(sql,d)

temp_df = pd.read_sql(sql,db)
temp_df.head(3)
```

```python
#  payment 테이블에서 월별 총매출. 평균매출. 렌탈 건수 합을 출력하시오
```

```python
sql="""
select * from payment;
"""

payment_df = pd.read_sql(sql,db)
payment_df['month'] = rental_df.rental_date.dt.month

grouped = rental_df.groupby('month')
```

```python
payment_df.head()
```

```python
grouped['amount'].mean()
```

```python
grouped.amount.mean()
```

```python
# groupby 객체 메소드 : agg, filter, transform

# agg : groupby 객체에 함수를 적용하는 메소드
# filter : 특정 group을 추출
# transform: 그룹의 값으로 데이터값을 바꿈
```

```python
payment_df.groupby('month').agg({'payment_id': np.size, 'amount': [np.sum, np.mean]})
```

```python
# sql

# left: 왼쪽에서부터 7자리 출력
sql= """
select count(payment_id) '구매건수', avg(amount) '평균구매액', sum(amount) '총구매액',
    left(payment_date, 7) as 'month'
    from payment
    group by month;
"""

payment_df = pd.read_sql(sql,db)
payment_df
```
