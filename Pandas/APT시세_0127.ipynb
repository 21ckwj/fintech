```python
#데이터 종류 확인

# print(df['분양가격(제곱미터)'].unique()) -> 전체 출력이 안됨

# sorted(df['분양가격(제곱미터)'].unique())  
# TypeError: '<' not supported between instances of 'float' and 'str'  -> 'float', 'str' 둘다 존재해서 정렬이 불가능

print(sorted(df['분양가격(제곱미터)'].astype(str).unique()))
```

```python
# 강사님 방법 - > 처음 float값이 아닌 것들 확인

indexes = []
errors= []

for i,j in enumerate(df['분양가격(제곱미터)']):
    try:
        float(j)
        indexes.append(i)
    except:
        errors.append(j)
errors
```

```python
# 에러값 전처리 : 문자열로 바꾸고 공백,',' 없애기

df['분양가격(제곱미터)'] = df['분양가격(제곱미터)'].apply(lambda x: str(x).strip().replace(',',''))
```

```python
# 다시 확인

indexes = []
errors= []

for i,j in enumerate(df['분양가격(제곱미터)']):
    try:
        float(j)
        indexes.append(i)
    except:
        errors.append(j)
errors
```

```python
# 에러가 아닌 값들만 다시 df_clean으로 정리

df_clean = df.iloc[indexes]
```

```python
df_clean['분양가격(제곱미터)'] = df_clean['분양가격(제곱미터)'].astype('float') * 3.3
```

```python
df_clean['평단분양가'] = df_clean['분양가격(제곱미터)']
df_clean.head()
```

```python
# 규모구분 column 전처리
```

```python
# 규모구분 데이터 확인

df_clean['규모구분'].unique()
```

```python
# df_clean['규모구분'] = df_clean['규모구분'].replace('전용면적','') 이건 데이터프레임에서 불가

# 정석은 apply 를 써야하지만 .str을 쓸 수도 있다

df_clean['규모구분'] = df_clean['규모구분'].str.replace('전용면적','')
df_clean['규모구분'] = df_clean['규모구분'].str.replace('제곱미터','')
df_clean['규모구분'] = df_clean['규모구분'].str.replace('이하','')
df_clean['규모구분'] = df_clean['규모구분'].str.replace('초과','~')
df_clean['규모구분'] = df_clean['규모구분'].str.strip()
df_clean['규모구분']
```

```python
df_clean.drop('분양가격(제곱미터)', axis=1, inplace= True)
df_clean
```

```python
dfc = df_clean.copy()
```
