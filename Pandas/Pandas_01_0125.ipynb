### missing value

```python
#matrix 형태를 dataframe으로

df = pd.DataFrame(np.random.randn(5,5), index=['a','b','c','d','e'], \
                  columns= ['one','two','three','four','five'])
df
```

```python
df.iloc[1:3,1] = np.nan  # iloc 함수는 파이썬 range[]와 같다
df.iloc[2, 2:4] = np.nan
df
```

```python
# Nan, NaN, nan 처리방식

# 0. nan값을 탐지 => isnull(), isna()
# 1. nan값을 포함한 row 또는 column을 삭제  => dropna()
# 2. nan값을 다른 값으로 채우는 방법 => fillna()
```

```python
df.isnull()  # Nan 값이면 True
```

```python
df.isnull().sum()  # column별 nan 개수
```

```python
df.isnull().sum(axis =1) # row별 nan개수
```

```python
df.isnull().sum().sum() #전체 nan개수
```

```python
#dropna()
```

```python
df.dropna()  # 기본으로 row를 삭제
```

```python
df.dropna(axis =1)
```

```python
df  # inplace= True 시 적용
```

```python
df.dropna(how='all')  # 전체 데이터가 nan인 경우 삭제
```

### merge/concat

* dataframe을 합치는 것.
* merge는 sql의 table을 합하는 join과 같다.

```python
from IPython.display import Image

Image('./Pandas_data_file/merge 사진.png')
```

```python
# merge의 종류

# A_df 와 B_df를 병합하는 경우
# 1.inner merge : A_df와 B_df 모두에 존재하는 데이터(row)들로만 합하는 방식
# 2.outer merge : A_df와 B_df 어느 한 곳에 존재하는 데이터를 모두 포함해서 합하는 방식
# 3.left merge : A_df의 데이터를 모두 포함. B_df에서 A_df에 존재하는 데이터만 포함
# 4.right merge : B_df의 데이터를 모두 포함. A_df에서 B_df에 존재하는 데이터만 포함
```

```python
raw_data = {
    'subject_id' : ['1','2','3','4','5','7','8','9','10','11'],
    'test_score' : [51,15,15,61,16,14,15,15,16,61]
}

df_a = pd.DataFrame(raw_data)
df_a
```

```python
raw_data = {
    'subject_id' : ['4','5','6','7','8'],
    'first_name' : ['Billy', 'Brian', 'Bran', 'Bryce', 'Betty'],
    'last_name'  : ['Bonder','Black', 'Balwner', 'Brice', 'Btisan']
}

df_b = pd.DataFrame(raw_data)
df_b
```

```python
df_a.subject_id.values
```

```python
df_b.subject_id.values
```

```python
merged_df = pd.merge(df_a,df_b,on='subject_id', how='inner') 
merged_df
```

```python
pd.merge(df_a,df_b, on='subject_id', how='outer') # outmerge : nan 값 생성 가능성 크다
```

```python
df_a.merge(df_b, on = 'subject_id', how='left')
```

```python
df_a.merge(df_b, on = 'subject_id', how='right')
```

### concat
