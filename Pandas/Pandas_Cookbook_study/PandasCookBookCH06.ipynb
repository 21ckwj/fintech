```python
# 소개 

- Series나 DataFrame의 모든 데이터 차원은 인덱스 객체에 레이블 되어있음
- pandas 사용자는 레이블을 사용해 데이터를 선택할 수 있고, 행과 열의 위치에 따라서도 데이터를 선택 할 수있음. 
- 이러한 레이블이나 위치별로 데이터를 선택하는것이 pandas에만 있는것은 아니지만, 
- 딕셔너리와 리스트에는 구체적인 지침이 있고 인덱스로 사용하는데 제약이있음.
- (예 - 딕셔너리키는 변경불가능한객체여야함,리스트는 선택을 위해 정수나 슬라이스 객체를 사용해야함.)
- pandas는 리스트처럼 정수를 사용하고 딕셔너리처럼 레이블을 사용해 데이터를 선택하는 기능을 병합함.
```

```python
# Series 데이터 선택 

- Series와 DataFrame은 인덱스 연산자 자체외에도 .iloc와 .loc 속성을 사용할 수 있고 고유한 방식으로 인덱스 연산자를 사용함.
- .loc, iloc : 스칼라, 값, 리스트, 슬라이스를 허용 
- .iloc 속성으로 인덱스 사용 => 위치별로 선택, 리스트와 유사하게 작동
- .loc 속성 => 인덱스 레이블로만 선택, 딕셔너리와 유사 
- 인덱스 연산: 변수[] , ex) s[item] , s.loc[item] 등과 같은 방식으로 데이터선택가능. 전자는 인덱스작업 직접수행, 후자는 .loc속성에서 작업
```

```python
## 작동 방법

#1. 단일열을 Series로 선택함
import pandas as pd
import numpy as np
college = pd.read_csv('data\9781784393878_Code\PandasCookbook_Code\data\college.csv', index_col='INSTNM')
city = college['CITY']
city
```

```python
# 2. .loc속성을 사용해 이름으로 스칼라값을 끄집어낸다
city.loc["Alabama A & M University"]
```

```python
#3. .iloc 속성을 사용해 위치로 스칼라 값을 끄집어 낸다.
city.iloc[0]
```

```python
#4. 인덱싱으로 여러값을 추출한다. 
# 인덱스 연산자에 리스트를 전달하면 pandas는 Series를 반환한다.
city[["Alabama A & M University","Alabama State University",]]
```

```python
# 5. 슬라이스를 .loc와 사용해 여러값을 추출 할 수 있음
city.loc[
    "Alabama A & M University":"Alabama State University"
]
```

```python
# 6. 슬라이스를 .iloc와 사용해 여러 위치로 여러값을 추출 할 수 있음
city.iloc[0:5]
```

```python
# 7. 불린 배열을 사용해 특정값을 추출한다 
alabama_mask = city.isin(["Birmingham","Montgomery"])
city[alabama_mask]
```

```python
# 주의할점 
# .iloc와 .loc는 명시적이지만 Series에서 직접 인덱싱하는것은 명시적이지 않기때문에 
# 되도록이면 Series를 직접 인덱싱 하지 않는다. 
```

### DataFrame 행 선택

```python
college = pd.read_csv('data\9781784393878_Code\PandasCookbook_Code\data\college.csv', index_col='INSTNM') 
#데이터셋불러와서 기관이름을 인덱스로 설정
college.sample(5, random_state=42)
```

```python
#그 위치에 있는 행 선택 
college.iloc[60] 
```

```python
# 이전과 동일한 행을 선택, 인덱스 레이블을 전달 
college.loc['University of Alaska Anchorage'] 
```

```python
# 불연속된 행 선택, .iloc에 정수 리스트 전달
# 행위치의 리스트를 전달했으므로 DataFrame을 반환한다
college.iloc[[60, 99, 3]] 
```

```python
# 전단계와 동일한 DataFrame 기관이름 복제
labels = ['University of Alaska Anchorage',
          'International Academy of Hair Design',
          'University of Alabama in Huntsville']
college.loc[labels] 
```

```python
# .iloc에 슬라이스표기 사용, 데이터에서 연속된 행을 선택 
college.iloc[99:102] 
```

```python
# 슬라이스 표기는 .loc에서도 사용가능 
start = "International Academy of Hair Design"
stop = "Mesa Community College"
college.loc[start:stop]
```

### DataFrame 행과 열을 동시에 선택 

```python
# # .loc와 .iloc로 행과 열을 선택하는 형태
# df.iloc[row_indxs, column_idxs]
# df.loc[row_names, column_names]
```

```python
college = pd.read_csv('data\9781784393878_Code\PandasCookbook_Code\data\college.csv', index_col='INSTNM')
college.iloc[:3, :4] # 데이터셋 읽고 처음 3개행, 4개열을 선택
# college.loc[:'Amridge University', :'MENONLY']
```

```python
college.iloc[90:80:-2,5] #행을 슬라이스하고 단일열을 선택
# start = 'Empire Beauty School-Flagstaff'
# stop = 'Arizona State University-Tempe'
# college.loc[start:stop:-2, 'RELAFFIL']
```

```python
# 행과 열을 동시에 선택하는 핵심 중 하나인 대괄호 안에 쉼표를 사용하는 방법을 이해한다.
# 쉼표 왼쪽은 항상 행 인덱스를 기준  ,오른쪽은 열 인덱스를 기준 
```

### 정수와 레이블 모두로 데이터 선택

```python
college = pd.read_csv('data\9781784393878_Code\PandasCookbook_Code\data\college.csv', index_col='INSTNM')
# 데이터셋 읽고 기관이름을 인덱스로 지정
```

```python
# 인덱서 메서드.get_loc을 사용해 원하는 열의 정수 위치를 찾음
col_start = college.columns.get_loc('UGDS_WHITE')
col_end = college.columns.get_loc('UGDS_UNKN') + 1
col_start, col_end
```

```python
college.iloc[:5, col_start:col_end] # iloc을 사용해 col_start와 col_end로 위치로 열을 선택함
```

```python
# 먼저 .get_loc메서드를 사용해 열 인덱스를 검색해서 슬라이스 하려는 열의 시작과 끝에 해당하는 정수 위치를 모두 찾음
# iloc를 사용한 슬라이싱은 마지막항목이 제외되므로 +1해준다
# 행과 열 위치에 슬라이스 표기법을 사용한다
```

### 사전식으로 슬라이싱

```python
#.loc 속성은 일반적으로 문자열레이블을 기반으로 데이터를 선택하지만 
# 인덱스 값의 사전식 순서에 따라 데이터를 선택할 수도 있다.
# 이 방법은 인덱스가 정렬되었을때만 작동한다. 
```

```python
college = pd.read_csv('data\9781784393878_Code\PandasCookbook_Code\data\college.csv', index_col='INSTNM')
# 데이터셋 읽고 기관이름을 인덱스로 지정
```

```python
college.loc['Sp':'Su'] # Sp와 Su사이에있는 이름을 가진 대학을 선택
#인덱스가 정렬되있지않아서 실패
```

```python
#먼저 인덱스를 정렬한다. 
college = college.sort_index()
```

```python
#다시 Sp와 Su사이에있는 이름을 가진 대학을 선택
college.loc['Sp':'Su']
```

```python
# 이 형식의 슬라이싱은 인덱스가 반대방향으로 정렬된 경우도 작동함
# .is_monotonic_increasing => 오름차순
# .is_monotonic_decreasing =>내림차순 
college = college.sort_index(ascending=False)
college.index.is_monotonic_decreasing
```
