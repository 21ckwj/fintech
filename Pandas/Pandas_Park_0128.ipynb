## 운동시설

```python
df1['공원보유시설(운동시설)'].value_counts()
```

```python
df1['공원보유시설(운동시설)'].unique().tolist()  # 문자열 안에 여러가지 시설이 들어가 있어서 정확한 파악어려움
```

```python
gym = df1['공원보유시설(운동시설)'].unique().tolist()
gym = str(gym)
```

```python
import re
```

```python
gym = gym.replace('+','.').replace('/','.').replace("''",'.').replace(' ','.') # 대입안해주면 안바뀐다
gym
```

```python
gym = re.sub("[\[\d\-nan\.\]]","",gym) # 맨밖 []: 사용자정의 정규표현식 
gym
```

```python
gym = re.sub(".{2.}", ".", gym)
gym
```

```python
# gym.split('.')
```

```python
pd.Series(gym.split(',')).value_counts()
```

```python
# pd.Series(gym.split('.')).value_counts().iloc[:30].plot(kind='bar', figsize=(10,5))
```

```python
def get_string(text):
    if pd.isna(text):
        return text
    
    text = re.sub("[\d\.'\-(nan)\[\]\/]","",text)
    text = re.sub(' ','',text)
    text = re.sub(".{2.}", ".", gym)
    return text
```

```python
df.columns
```

```python
df['편익시설'] = df['공원보유시설(편익시설)'].apply(get_string)
```
