#### 7자리 이상의 임의의 정수를 입력 받은 후, 1000 단위 구별자를 추가하여 리턴하는 코드를 작성하시오
#### ex) 123456789 =>123,456,789

```python
# 강사님 코드

num = input("임의의 7자리 이상인 정수를 입력하세요:")

num = num[::-1]  # 123456789 -> 987654321 인덱스 0부터 쉼표 찍어도 된다
ret= ''

for i,c in enumerate(num):
    i+=1
    if (i%3 ==0) & (i != len(num)): # 3의배수 + 마지막 자리는 아닌 i
        ret += c + ','              # ret 이라는 빈 문자열에 value 값들을 집어넣어라 
    else:
        ret += c
ret = ret[::-1]
print(ret)
```
