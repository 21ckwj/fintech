```python
# path = '#old_content > table > tbody > tr > td.title'  #박스 (경로 수정해야 데이터 긁어와짐)
path = 'tbody > tr > td.title'  
tags = soup.select(path)
tags
```

```python
# 제우 형 코드
import requests
from bs4 import BeautifulSoup

baseurl = 'https://movie.naver.com/movie/point/af/list.naver?&page='
for page in range(1, 11):
    res = requests.get(baseurl+str(page))
    soup = BeautifulSoup(res.text,'html.parser')
    #path = '#old_content > table.list_netizen > colgroup > tbody > tr > td.title'
# 이 망할 구조를 모르면 상위 단계를 짤라보는 것도 방법이랜다.    
    path = 'tbody > tr > td.title'
    ratinglst = soup.select(path)

    for lst in ratinglst:
        title = lst.select('a')[0].text
        print('영화제목 :',title)
        rating = lst.select('div.list_netizen_score > em')[0].text
        
        print('평점 :',rating)
        
        content = str(lst)
        print('감상평 :' ,content[content.index('<br/>')+5:content.index('<a class="report"')].strip(' \t\n'))
        #print('감상평 :' ,' '.join(content[content.index('<br/>')+5:content.index('<a class="report"')].split()))
```
