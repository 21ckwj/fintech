```python
# 7.sakila 데이터베이스의 rental 테이블에서 월별 렌탈 수를 출력하시오
```

```python
sql ="use sakila;"
cursor.execute(sql)
```

```python
sql= """
select * from rental;"""

df = pd.read_sql(sql, db)
df.head()
```

```python
# sakila 데이터베이스의 rental 테이블에서 월별 렌탈 수를 출력하시오
# (1)  pandas

# 월별을 추출하기 위해 rental_date 의 데이터 형식 확인
df.info() # datetime 형식
```

```python
# rental_date 열에서 년,월 추출
import datetime

df['year-month'] = df['rental_date'].dt.strftime("%Y-%m")
df['year-month']
```

```python
df.groupby('year-month')['rental_date'].count()
```

```python
# sakila 데이터베이스의 rental 테이블에서 월별 렌탈 수를 출력하시오
# (2) sql
df
```

```python
# sql 사용
sql = """
select left(rental_date,7),count(rental_id) from rental
group by left(rental_date,7);
"""

pd.read_sql(sql,db)
```

```python
# 8.payment 테이블에서 월별 총매출. 평균매출. 렌탈 건수 합을 출력하시오
```

```python
sql ="""select * from payment"""

df =pd.read_sql(sql,db)
df.head()
```

```python
# 데이터 타입 확인
df.info()
```

```python
df['month'] = df.payment_date.dt.strftime('%m')
df['month']
```

```python
#월별 총 매출
df.groupby('month').sum('amount')['amount']
```

```python
# 평균매출
df.groupby('month').mean('amount')['amount']
```

```python
# 렌탈 건수 합
df.groupby('month').count()['amount']
```
