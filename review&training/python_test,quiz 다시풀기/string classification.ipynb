### 특수문자, 공백 빼기
#### 1. 처음부터 전처리 후 세기
#### 2. 카운트 끝나고 마지막에 제거하기

```python
# 1. 처음부터 전처리 후 세기
import string
f = open("test1_0112.txt" ,'r', encoding="UTF-8")   
line = f.read()

lst=[]
# 비어있는 리스트에 특수문자 아닌 것을 모으자
for c in line:
    if c not in string.punctuation:
        lst.append(c)
# print(lst,end='')       
# '\n' 지우기 (특수문자에 포함 안된 기호)  리스트의 특정값 모두 지우는 다른 방법은 없을까?
for c in line:
    if c in '\n':
        lst.remove(c)
# print(lst,end='')

char = {}

for c in lst:
    if c not in char:
        char[c]=1
    else:
        char[c]+=1
        
for k,v in char.items():
    print("{} 의 등장횟수 = {}".format(k,v))
```

```python
# 2. 카운트 끝나고 마지막에 제거하기
import string

f = open("test1_0112.txt",'r',encoding= 'UTF-8')
line = f.read()
line
char = {}
for c in line:
    if c not in char:
        char[c] =1
    else:
        char[c]+=1

# 특수문자 제거를 하고 다시 clena_char 딕셔너리에 모아줌
clean_char = {}

for k,v in char.items():
    if k not in string.punctuation:
        clean_char[k] = v
# 오름차순 정렬
# clean_char = sorted(clean_char.items(), reverse=True)  -> 리스트 형태로 바뀜 
clean_char = dict(sorted(clean_char.items(), reverse=False))
clean_char
# 출력하기
for k,v in clean_char.items():
    print("{} 등장횟수 => {}".format(k,v))
```
