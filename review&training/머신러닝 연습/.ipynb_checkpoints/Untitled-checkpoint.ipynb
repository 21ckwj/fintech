## 해보기

```python
# 입원했던 환자가 30일 이후에 재입원할 확률
# 입원 visit_occu_id = 9201
df_vo.info()
```

```python
# 재입원했던 환자들을 추려야함
df_vo_h = df_vo[df_vo['visit_concept_id']==9201]
df_vo_h
```

```python
df_vo_h['visit_start_datetime'] = pd.to_datetime(df_vo_h['visit_start_datetime'])
df_vo_h['visit_end_datetime'] = pd.to_datetime(df_vo_h['visit_end_datetime'])
```

```python
df_vo_h.info()
```

```python
# 재입원환자 목록
df_vo_h['person_id'].value_counts()
```

```python
# person_id -> visit_start_date 정렬

df_vo_h.sort_values(by=['person_id','visit_start_datetime'],ascending=[False, True]) 
```
