```python
# newwork페이지에 나온 정보

url = 'http://www.cine21.com/rank/person/content'
```

```python
date = '2018-08'

data= {'section': 'actor',
'period_start': date,
'gender': 'all',
'page': 1}

```

```python
res = requests.post(url,data = data)
res
```

```python
soup = BeautifulSoup(res.text, 'html.parser')
soup
```

```python
# 한칸의 경로가 안긁어짐
path = '#rank_holder > ul > li.people_li'
lis = soup.select(path)
lis
```

```python
# 배우이름 + 배우 url

path =  'ul > li.people_li > div.name'
lis = soup.select(path)
lis
```

```python
# 배우이름

lis[0].text
```

```python
# 배우 url 뽑기
# lis 리스트 벗겨주고 ->  <div class=''><a href=''> 중 a태그선택 -> 리스트 벗겨주고 -> 'href' 선택

print(lis[0].select('a')[0]['href'])

main_url = 'http://www.cine21.com'
actor_url = main_url + lis[0].select('a')[0]['href']

actor_url
```

```python
for li in lis:
    print(li.text, actor_url )
```
