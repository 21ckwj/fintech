## 단어를 숫자로 전환

### 1. counter vectorizing

        형태소1, 형태소2, 형태소3, 형태소4 ...
문서1     0        2        1        0   =>  (0,2,1,0  ...)<br>
문서2     1        1        0        0           =>  (1,1,0,0, ...)<br>
문서3     0        0        3        0           =>  (0,0,3,0, ...)<br>
...

### 2. tfidf vectorizing

문서에서 형태소1 등장수 : tf<br>
전체 문서에서 형태소1이 등장한 문서수 : idf<br>
tfidf = tf/idf => 특정 형태소가 특정 문서에서의 중요도<br>

        형태소1, 형태소2, 형태소3, 형태소4 ...
문서1     0        0.3      0.01     0           =>  ( 0, 0.3 , 0.01, 0  ...)<br>
문서2     0.005    0.07     0        0           =>  (0.005, 0.07,0,0, ...)<br>
문서3     0        0        0.35     0           =>  (0,0,0.35,0, ...)<br>

```python
for i in range(len(docs)):
    print(docs[i])
```

```python
# 텍스트 형식 예시 
L =['나라', '대한', '민국']
' '.join(L)
```

```python
# CountVectorizer 사용시 텍스트 형식으로 들어가야함
for i in range(len(docs)):
    docs[i] = ' '.join(docs[i])
```

```python
docs
```

```python
vec = CountVectorizer()
x = vec.fit_transform(docs)
```

```python
news_arr = x.toarray()
news_arr
```

```python
news_arr.shape
```

```python
news_arr[0] # sparse matrix => 학습이 잘 안됨
```

```python
vec.get_feature_names()  
```

```python
news_df = pd.DataFrame(news_arr, columns =vec.get_feature_names() )
news_df
```

## clustering

```python
 from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=3)
kmeans.fit(news_df)
```

```python
# 정답값 : 00000(유한양행) 11111(라돈침대) 22222(복면가왕) 숫자는 의미없음 잘 묶였는지가 중요!
```

```python
kmeans.labels_
```

```python
cluster = AgglomerativeClustering(n_clusters=3, linkage='ward')
cluster.fit_predict(news_df)
```

```python
cluster = DBSCAN(eps=3, min_samples=5) # 특정거리내에 밀도 높은거 묶는지
cluster.fit_predict(news_df)
```

### 차원축소 사용

### 크롤링 완성코드

```python
url_eco = 'https://news.daum.net/breakingnews/economic?page=1'
res = requests.get(url_eco)
soup = BeautifulSoup(res.text, 'html.parser')
eco_lst = soup.select('#mArticle > div.box_etc > ul> li')
eco_lst
```

```python
# 카테고리
category = ['economic', 'entertain', 'politics']
cate_news_lst= []

for c in category:
    url_eco = 'https://news.daum.net/breakingnews/{}?page=1'.format(c)
    res = requests.get(url_eco)
    soup = BeautifulSoup(res.text, 'html.parser')
    eco_lst = soup.select('#mArticle > div.box_etc > ul> li')

    title_lst=[]
    summary_lst = []
    news_lst= []
    news_lst1 = []

    for eco in eco_lst:
        title = eco.select('li > div > strong > a')[0].text
        summary = eco.select('ul > li > div > div')[0].text
        title_lst.append(title)
        summary_lst.append(summary)

        print(title)
        print(summary)


        #전체뉴스
        news_url = eco.select('li > div > strong > a')[0]['href']
        res1 = requests.get(news_url)
        soup1 = BeautifulSoup(res1.text,'html.parser')
        lines = soup1.select('#harmonyContainer > section> p') # p태그들을 모은 값

        news=''

        for line in lines:
            new1 = line.text    # 기사 한줄
            news += new1 # 전체 기사
        print(news)
        news_lst.append(news)
        print('-'*50) 
        
    cate_news_lst.append(news_lst)
```
