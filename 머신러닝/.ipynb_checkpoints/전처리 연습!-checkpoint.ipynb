```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("./data_ml/titanic_train.csv")
df.head(3)
```

```python
# male = 0 femal = 1 로 바꾸기

df['Sex'] = df['Sex'].str.replace('female','0')
df['Sex'] = df['Sex'].str.replace('male', '1')
df['Sex'] = df['Sex'].astype(float)
df['Sex']
```

```python
## 원중이 바보 by 류제우
df['Sex'].apply(lambda x : 0 if x == 'male' else 1)
```
