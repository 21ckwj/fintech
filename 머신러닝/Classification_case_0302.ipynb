```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('./data_ml/santander_train.csv')
df.head()
```

```python
# Target => 불만족:1 , 만족:0
```

```python
df.TARGET.value_counts()
```

```python
df.describe() #var3 의 최솟값 고쳐줘야함
```

```python
# 이상치(outlier) 제거

df.var3.value_counts().sort_index()
df['var3'].replace(-999999,2, inplace=True)
```

```python
 # x,y date divide
```

```python
y= df['TARGET']
x= df.drop('TARGET', axis=1)
```

```python
x.drop(columns= ['ID'], inplace=True)
```

```python
x.head(2)
```

```python
# train test split
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.3, random_state=111)
```

```python
# algorithm import
```

```python
from xgboost import XGBClassifier
```

```python
xgb_clf = XGBClassifier(n_estimators =500, random_state=111)
xgb_clf.fit(x_train,y_train, early_stopping_rounds=100, eval_metric='auc',eval_set=[(x_train,y_train),(x_test,y_test)])
# 100번까지 기다렸다가 더이상 나이지않으면 중단
```

```python
#model evaluation
```

```python
from sklearn.metrics import classification_report, roc_auc_score
```

```python
preds = xgb_clf.predict(x_test)
```

```python
print(classification_report(y_test, preds))
```

```python
roc_auc_score(y_test, xgb_clf.predict_proba(x_test)[:,1], average='macro')
```

```python
# 어떤 파라미터인지 정보 없을때 : 구글검색해서 캐글 혹은 유명한사람 코드 찾아서~
```

```python
from sklearn.model_selection import GridSearchCV

xgb_cls = XGBClassifier(n_estimators=100)
params = {'max_depth':[5,7], 'min_child_weight':[1,3], 'colsmaple_bytree': [0.5,0.7]}

grid = GridSearchCV(xgb_cls, param_grid= params, verbose= False)
grid.fit(x_train,y_train, early_stopping_rounds=30, eval_metric='auc',eval_set=[(x_train,y_train),(x_test,y_test)])
```

```python
grid.best_params_
```

```python
grid.best_score_
```
