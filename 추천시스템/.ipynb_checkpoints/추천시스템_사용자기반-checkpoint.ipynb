# 사용자 기반

```python
merge_df = pd.merge(movies, ratings, on='movieId')
merge_df
```

```python
merge_df.info()
```

```python
ratings['movieId'] = ratings['movieId'].astype('int16')
ratings['userId'] = ratings['userId'].astype('int16')
ratings['rating'] = ratings['rating'].astype('int16')
```

```python
ratings = merge_df[['movieId','userId','rating']]
rating_matrix = ratings.pivot_table(index = 'userId', columns = 'movieId', values='rating')
```

## 메모리 문제로 피클파일로 대체

```python
import pickle
with open('./data/ratings_matrix.pickle','rb') as f:
    ratings_matrix = pickle.load(f)
```

```python
# 사용자평점 매트릭스 - 사용자 기반 추천 시스템
# sparse matrix - 사용자가 직접 본 영화의 수는 제한적이다

ratings_matrix
```

```python
ratings_matrix = ratings_matrix.fillna(0)
ratings_matrix.head()
```

```python
# 고객님과 유사한 사용자가 다음의 콘텐츠도 구매하였습니다.
# 로맨스물의 평점이 낫다면 
```

```python
from sklearn.metrics.pairwise import cosine_similarity

user_sim = cosine_similarity(ratings_matrix)
user_sim
```

```python
user_sim.shape
```

```python
user_sim_sort = user_sim.argsort(axis=1)[:,::-1]
user_sim_sort
```

```python
user_sim[0].argsort()[::-1]
```

```python
user_sim_sort[10][1:11] # 유사도가 높은 사용자 상위 10명
```

```python
user_sim_sort[10]
```

```python
# 유사도가 높은 사용자 상위 10명의 영화 평점 데이터
df = ratings_matrix.loc[user_sim_sort[10][1:11]]
df
```

```python
# 10번 사용자와 유사도가 높은 사용자들의 영화별 평점합의 순위

df.sum().sort_values(ascending =False)[:5]
```

## 함수화

```python
import pickle
with open('./data/ratings_matrix.pickle','rb') as f:
    ratings_matrix = pickle.load(f)
    
ratings_matrix = ratings_matrix.fillna(0)
    
def sim_user_based_recomm(ratings_matrix,user_id, n_top=10):
    user_sim = cosine_similarity(ratings_matrix)
    user_sim_sort = user_sim[user_id].argsort()[::-1]
    df = ratings_matrix.loc[user_sim_sort[1:11]]
    return df.sum().sort_values(ascending =False)[:5]
```

```python
sim_user_based_recomm(ratings_matrix,30 )
```

## 함수화2- 봤던 영화 제외

```python
def seen_movies(user_id):
    movie_list = ratings_matrix.iloc[user_id]
    seen_movies = movie_list[ratings_matrix.iloc[user_id]>0]
    return seen_movies.index
    
```

```python
seen_movies(30)
```

```python
import pickle
with open('./data/ratings_matrix.pickle','rb') as f:
    ratings_matrix = pickle.load(f)
    
ratings_matrix = ratings_matrix.fillna(0)
    
def sim_user_based_recomm(ratings_matrix,user_id, n_top=10):
    seen_movie_list = seen_movies(user_id)
    user_sim = cosine_similarity(ratings_matrix)
    user_sim_sort = user_sim[user_id].argsort()[::-1][:n_top*5]
    df = ratings_matrix.loc[user_sim_sort]
    
    unseen_movie_list = set(df.index).difference(set(seen_movie_list))
    
    df= df.loc[unseen_movie_list]
    
    return df.sum().sort_values(ascending =False)[:5]
```

```python
sim_user_based_recomm(ratings_matrix, 30, n_top=10)
```

# 아이템 기반 이웃 협업 필터링

```python
# 사용자 평점이 유사한 영화들 간의 유사도 계산
```

```python
ratings_matrix.head()
```

```python
item_sim = cosine_similarity(ratings_matrix.T, ratings_matrix.T)
item_sim
```

```python
item_sim.shape
```

```python
item_sim_df = pd.DataFrame(item_sim, index=ratings_matrix.columns, columns = ratings_matrix.columns)
item_sim_df.head(3)
```

```python
item_sim_df['Godfather, The (1972)'].sort_values(ascending=False)[:10]
```

```python
item_sim_df['American Beauty (1999)'].sort_values(ascending=False)[:10]
```

# 영화의 유사도와 사용자 평점을 함께 고려한 영화 추천

```python
ratings_matrix.head()
```

```python
ratings_matrix.mean()
```

```python
# 영화의 유사도 점수에 해당 영화의 평점을 곱한 것

item_co_pro = item_sim_df * (ratings_matrix.mean())
item_co_pro.head()
```

```python
item_co_pro['Godfather, The (1972)'].sort_values(ascending=False)[:10]
```

### 개인화된 아이템 기반 협업 필터링

```python
# user의 평점점수와 영화의 유사도 점수를 곱하고 더해서 나온 값
# user가 높은 평점을 준 영화에 대해서 특정 영화가 높은 유사도를 가지면 높은 점수가 부여
# 큰값(유저의 특정 영화에 대한 평점)과 큰값(해당 영화가 높은평점의 영화와 유사도)이 곱해져 큰값 나온다

ratings_matrix.dot(item_sim_df)
```
