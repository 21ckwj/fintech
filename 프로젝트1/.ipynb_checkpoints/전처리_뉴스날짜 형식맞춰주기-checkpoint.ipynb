```python
last_date = datetime(2022,2,10)

for i in range(len(df['날짜1'])):
    cnt = len(df['날짜1'][i])
    
    if cnt <=2:
        day_c = int(df['날짜1'][i])
        date_f = last_date + timedelta(days= -day_c)
        date_f = str(date_f)
    else:
        date_f = df['날짜1'][i]
    
    df['날짜'][i] =date_f[:10]
    
    
    
# df['날짜1'].unique()

# datetime.datetime.strptime("21:00:00", "%H:%M:%S") + datetime.timedelta(hours=2)
```
