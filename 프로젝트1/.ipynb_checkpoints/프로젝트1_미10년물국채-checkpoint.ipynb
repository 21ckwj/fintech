```python
import pandas as pd
import numpy as np
```

```python
import plotly.express as px
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.graph_objects as go
```

```python
from datetime import datetime, timedelta
```

```python
## 미 10년물 국채 금리

df = pd.read_csv('./data/미10년물 국채 금리.csv')
df
```

```python
df.info()
```

```python
df['날짜'] = df['날짜'].str.replace('년','-')
df['날짜'] = df['날짜'].str.replace('월','-')
df['날짜'] = df['날짜'].str.replace('일','')
df['날짜'] = df['날짜'].str.replace(' ','')
df['날짜']
```

```python
df['날짜']= pd.to_datetime(df['날짜'])
```

```python
df_10= df.set_index('날짜').loc['2017-2-10':'2022-02-10'].resample('W-MON',label = 'left', closed='right')['종가'].agg(['max','min','first','last'])
df_10
```

```python
## 원유재고
```

```python
df1 = pd.read_csv('./data/미국원유재고1.csv',delimiter='\t')
df1.head(3)
```

```python
df1['발표일'] = df1['발표일'].str.replace('년','-')
df1['발표일'] = df1['발표일'].str.replace('월','-')
df1['발표일'] = df1['발표일'].str.replace('일','')
df1['발표일'] = df1['발표일'].str.replace(' ','')
df1['발표일'] = pd.to_datetime(df1['발표일'])
df1['발표일']
```

```python
df1 = df1.rename(columns= {'발표일':'날짜','발표':'미국원유재고량'})
df1 = df1.drop(columns = ['시간','예측','이전','Unnamed: 5'])
df1.head(3)
```

```python
df1
```

```python
df_10= df1.set_index('날짜').loc['2017-2-10':'2022-02-10'].resample('W-MON',label = 'left', closed='right')['미국원유재고량'].agg(['max','min','first','last'])
df_10
```
