```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# ◆ 산불_피해면적기준

```python
df_sanbul = pd.read_csv('./data/산불타겟추가_피해면적기준.csv',index_col=0)
df_sanbul = df_sanbul.set_index('시간')
df_sanbul
```

```python
# 2019년은 test를 위해 빼놓음

df_sanbul1 = df_sanbul.loc['2003-01-01':'2018-12-31']
df_sanbul2 = df_sanbul.loc['2020-01-01':'2021-12-31']

df_sanbul3 = pd.concat([df_sanbul1,df_sanbul2])
df_sanbul3
```

## (1) 일주일 선행

```python
y = df_sanbul3['산불발생여부'][:-7]
x = df_sanbul3.drop(columns = ['산불발생여부']).shift(7)['2003-01-08':]
```

```python
y[:-7]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=111)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC

from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import GradientBoostingClassifier


from xgboost import XGBClassifier
# from lightgbm import LGBMClassifier
from catboost import CatBoostClassifier
```

```python
from sklearn.metrics import classification_report, f1_score, roc_auc_score

log = LogisticRegression()
knn = KNeighborsClassifier()
svc = SVC(probability=True)
rfc = RandomForestClassifier()
ada = AdaBoostClassifier()
gbm = GradientBoostingClassifier()

xgb = XGBClassifier()
# lgbm = LGBMClassifier()
cat = CatBoostClassifier()

# model1 = [svc, ada, gbm]
# model2 = [rfc, xgb, lgbm, cat]
model = [log, knn, svc, rfc, ada, gbm,xgb, cat]
```

```python
model = [log, knn, svc, rfc, ada, gbm,xgb, cat]
for m in model:
    m.fit(x_train,y_train)
    preds = m.predict(x_test)
    
    print(m.__class__.__name__)
    print(classification_report(y_test, preds))
    print(roc_auc_score(y_test,m.predict_proba(x_test)[:,1]))
    print('-'* 50)
```

## scaler 써보기 

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
scaler = StandardScaler()
scaler.fit(x_train)

x_train_scaled = scaler.transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
x_test_scaled
```

```python
# model = [svc, rfc, ada, gbm,xgb, cat]

for m in model:
    m.fit(x_train_scaled,y_train)
    preds = m.predict(x_test_scaled)
    
    print(m.__class__.__name__)
    print(classification_report(y_test, preds))
    print(roc_auc_score(y_test,m.predict_proba(x_test_scaled)[:,1]))
    print('-'* 50)
```
