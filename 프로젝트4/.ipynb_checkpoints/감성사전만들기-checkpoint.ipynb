```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from bs4 import BeautifulSoup
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
```

```python
# 날짜 전처리
import re
df = pd.read_csv('./data/politics/news_politics_00001.csv',index_col=0).dropna()
df['date'] = df['date'].astype(str).str.strip('<span>').str.strip('</span>')
df['date'] = df['date'].str.replace('\n','').str.replace('—','')
df['date'] = df['date'].apply(lambda x: re.findall('\(([^)]+)',x)[0] if re.findall('\(([^)]+)',x) else x )
df['date'] = pd.to_datetime(df['date'].astype(str),utc=True).astype(str).apply(lambda x: x[:10]).astype('datetime64')
```

```python
df.head()
```

```python
# 전처리 
def preprocessing(news):
    
    # HTML 태그 제거
    news_text = BeautifulSoup(news, "html5lib").get_text()
    
    # 영어가 아닌 특수문자를 공백(" ")으로 바꾸기
    news_text = re.sub("[^a-zA-Z]"," ", news_text)
    
    # 대문자를 소문자로 바꾸고 공백 단위로 텍스트를 나누어 리스트화
    words = news_text.lower().split()
    
    # 어원화
    stemmer = PorterStemmer()
    words = [stemmer.stem(w) for w in words]
    
    # 불용어 제거하기
    stop_words = stopwords.words('english')
    words = [w for w in words if not w in stop_words]
    
    # 한글자 없애기
    words = [w for w in words if len(w)>1]
    
    clean_news = ' '.join(words)
        
    clean_list = clean_news.split()
    clean_list.extend(' ')
    
    return clean_list
```

```python
df['clean_news'] = df['content'].apply(preprocessing)
df['title'] = df['title'].apply(preprocessing)
```

```python
df1 = df.sort_values(by='date', ascending = True)
df1.head()
```

```python
df1.reset_index(drop=True)
```

```python
df_date = df1.groupby('date')[['title','clean_news']].sum()
df_date.reset_index(inplace=True)
df_date.head()
```

```python
df_oil = pd.read_csv('./data/석유와가스_지수.csv',index_col=0)
df_oil.head()
```

```python
# merge를 위한 칼럼이름 변경
df_oil = df_oil.rename(columns={'Date':'date'})
df_oil.head()
```

```python
df_oil['date'] = df_oil['date'].astype('datetime64')
```

```python
df_oil_idx = df_oil[['date','전일비']]
df_oil_idx.head()
```

```python
# 날짜 하루 뒤 반영? 나중에 생각

df_merge = pd.merge(df_date,df_oil_idx,how='inner')
df_merge1 = df_merge[['date', 'title','clean_news','전일비']]
```

```python
df_merge1.head()
```

```python
df_merge1[df_merge1['전일비']< 0] # 120
df_merge1[df_merge1['전일비']> 0] # 128
```

```python
senti_dic = dict()

for d in range(len(df_merge1['date'])):  # 날짜
    
    # 하루뉴스의 단어 빈도수 series
    words_cnt = pd.Series(df_merge1['clean_news'][d]).value_counts()   

    # 단어이름
    words = words_cnt.index

    for word in words:

        #딕셔너리에 없는 단어일 경우 (처음 빈도수값*변동률) 대입
        if word not in senti_dic:
            senti_dic[word] = words_cnt.loc[word] * df_merge['전일비'][d]

        # 기존에 있는 단어라면 기존꺼에 추가
        else:
            senti_dic[word] += words_cnt.loc[word] * df_merge['전일비'][d]

senti_dic    
```

```python
senti_dic
```

```python
# 감성점수기반 뉴스의 점수
senti_lst = []

for d in range(len(df_merge1['date'])):
    news_words = df_merge1['clean_news'][d]

    senti_score = 0
    for word in news_words:
        if word in senti_dic.keys():
            senti_score += senti_dic[word]
        else:
            continue
    
    senti_lst.append(senti_score)
        
senti_lst
```

```python
# 날짜별 감성점수
plt.plot(senti_lst)
```
