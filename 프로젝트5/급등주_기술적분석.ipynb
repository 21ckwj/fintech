## 고려산업

```python
df_go = fdr.DataReader('002140')
df_go
```

```python
## 60일선,120일선
```

```python
df_go['ma20'] = df_go['Close'].rolling(20).mean()
df_go['ma60'] = df_go['Close'].rolling(60).mean()
df_go['ma120'] = df_go['Close'].rolling(120).mean()
```

```python
# 20일 표준편차
df_go['std20'] = df_go['Close'].rolling(20).std()
```

```python
# 볼린져 밴드
df_go['upper_b'] = df_go['ma20'] + (df_go['std20']*2)
df_go['lower_b'] = df_go['ma20'] - (df_go['std20']*2)
```

```python
df_go.dropna(inplace=True)
```
